```python
# prompt: load the data from https://github.com/geoffswc/Document-Classification/raw/refs/heads/master/data/reviews.csv into a pandas dataframe

import pandas as pd

# Load the dataframe
df = pd.read_csv('https://github.com/geoffswc/Document-Classification/raw/refs/heads/master/data/reviews.csv')

```

```python
# prompt: Using dataframe df: use the VADER open source app to generate sentiment analysis scores for every text item

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment_scores(text):
    return analyzer.polarity_scores(text)

# Apply sentiment analysis to the 'text' column
df['sentiment_scores'] = df['text'].apply(get_sentiment_scores)
```

```python
df.head()
```

```python
# prompt: Using dataframe df: create a new column vader_review with a 1 or 0 value based on whether the pos or neg score in sentiment_scores is higher.

import pandas as pd

# Assuming 'sentiment_scores' is a column containing dictionaries
# and each dictionary has keys 'pos' and 'neg'.

# Convert the sentiment_scores column back to a dictionary if it is string
#df['sentiment_scores'] = df['sentiment_scores'].apply(eval)

# Create a new column 'vader_review' based on positive and negative scores.
df['vader_review'] = df['sentiment_scores'].apply(lambda x: 1 if x['pos'] > x['neg'] else 0)
```

```python
df[['text', 'review', 'vader_review']]
```

```python
# prompt: calculate the precision, recall, and accuracy for the vader_review

from sklearn.metrics import precision_score, recall_score, accuracy_score

# Calculate precision, recall, and accuracy
precision = precision_score(df['review'], df['vader_review'])
recall = recall_score(df['review'], df['vader_review'])
accuracy = accuracy_score(df['review'], df['vader_review'])

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")
```
